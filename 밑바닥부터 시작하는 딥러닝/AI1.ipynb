AND 퍼셉트론

```python
def AND(x1,x2):
  w1,w2,theta = 0.5,0.5,0.7
  tmp = x1*w1 + x2*w2
  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1

x = np.array([0,1])
w = np.array([0.5,0.5])

b = -0.7

w*x
```

```python
# 입력 데이터 생성
x1 = np.linspace(0, 1, 100)  # x1 값 범위 (0에서 1까지)
x2 = np.linspace(0, 1, 100)  # x2 값 범위 (0에서 1까지)

X1, X2 = np.meshgrid(x1, x2)  # 그리드 생성
Z = np.array([[AND(x1, x2) for x1, x2 in zip(row_x1, row_x2)]
              for row_x1, row_x2 in zip(X1, X2)])  # AND 결과값 계산

# 경사도(결정 경계) 시각화
plt.contourf(
    X1,
    X2,
    Z,
    levels=[0, 0.5, 1.0],
    colors=['blue', 'red'],
    alpha=0.3
)

# 데이터 포인트 추가 (AND 게이트의 4가지 조합)
data_x1 = [0, 0, 1, 1]
data_x2 = [0, 1, 0, 1]
data_y = [AND(a, b) for a, b in zip(data_x1, data_x2)]

# 점 그리기
for i in range(len(data_x1)):
    color = 'red' if data_y[i] == 1 else 'blue'
    plt.scatter(data_x1[i], data_x2[i], color=color, s=100)

# 그래프 꾸미기
plt.title("AND Gate Decision Boundary")
plt.xlabel("Input x1")
plt.ylabel("Input x2")
plt.grid(True)
plt.show()
plt.show()
```

NAND 퍼셉트론

```python
def NAND(x1,x2):
  x = np.array([x1,x2])
  w = np.array([-0.5,-0.5])
  b = 0.7
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

```python
# 입력 데이터 생성
x1 = np.linspace(0, 1, 100)  # x1 값 범위 (0에서 1까지)
x2 = np.linspace(0, 1, 100)  # x2 값 범위 (0에서 1까지)

X1, X2 = np.meshgrid(x1, x2)  # 그리드 생성
Z = np.array([[NAND(x1, x2) for x1, x2 in zip(row_x1, row_x2)]
              for row_x1, row_x2 in zip(X1, X2)])  # AND 결과값 계산

# 경사도(결정 경계) 시각화
plt.contourf(
    X1,
    X2,
    Z,
    levels=[0, 0.5, 1.0],
    colors=['red', 'blue'],
    alpha=0.3
)

# 데이터 포인트 추가 (AND 게이트의 4가지 조합)
data_x1 = [0, 0, 1, 1]
data_x2 = [0, 1, 0, 1]
data_y = [NAND(a, b) for a, b in zip(data_x1, data_x2)]

# 점 그리기
for i in range(len(data_x1)):
    color = 'blue' if data_y[i] == 1 else 'red'
    plt.scatter(data_x1[i], data_x2[i], color=color, s=100)

# 그래프 꾸미기
plt.title("NAND Gate Decision Boundary")
plt.xlabel("Input x1")
plt.ylabel("Input x2")
plt.grid(True)
plt.show()
plt.show()
```

OR 퍼셉트론

```python
def OR(x1,x2):
  x = np.array([x1,x2])
  w = np.array([0.5,0.5])
  b = -0.2
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```
